```python
### Fine grid values 

def calculate_statistics(base_path, varDir, nseeds):
    """Calculate statistics for fine grid values across different connectivity structures and variances."""
    
    k_arr = np.array([])
    kg_arr = np.array([])
    var_arr = np.array([])
    
    for ivar in range(len(varDir)):
        rdir = varDir[ivar]
        k1 = np.array([])
        kg1 = np.array([])
        var1 = np.array([])
        
        for iseed in range(nseeds):
            try:
                filepath = f"{base_path}{rdir}{iseed}/k.npy"
                aux = np.load(filepath).reshape(-1)
                
                k1 = np.append(k1, np.mean(aux))
                kg1 = np.append(kg1, np.exp(np.mean(np.log(aux))))
                var1 = np.append(var1, np.var(np.log(aux)))
            except Exception as e:
                print(f"Error processing {filepath}: {e}")
                continue
                
        k_arr = np.append(k_arr, np.mean(k1))
        kg_arr = np.append(kg_arr, np.mean(kg1))
        var_arr = np.append(var_arr, np.mean(var1))
        
    return (np.reshape(k_arr, (1,len(varDir))), 
            np.reshape(kg_arr, (1,len(varDir))), 
            np.reshape(var_arr, (1,len(varDir))))

# Calculate statistics for each connectivity type
connectivity_types = {
    'con': './3d/3dcon/',
    'int': './3d/3dint/',
    'dis': './3d/3ddis/'
}
results = {}

for conn_type, base_path in connectivity_types.items():
    k, kg, var = calculate_statistics(base_path, varDir, nseeds)
    results[conn_type] = {
        'k': k,
        'kg': kg,
        'var': var
    }
    print(f"\nResults for {conn_type}:")
    print(f"k: {k}, shape: {k.shape}")
    print(f"kg: {kg}, shape: {kg.shape}")
    print(f"var: {var}, shape: {var.shape}")

# Save results to .npz file
save_path = '3D_data/3D_fine_grid_stats.npz'
try:
    np.savez(save_path,
             k_con=results['con']['k'],
             kg_con=results['con']['kg'],
             var_con=results['con']['var'],
             k_int=results['int']['k'],
             kg_int=results['int']['kg'],
             var_int=results['int']['var'],
             k_dis=results['dis']['k'],
             kg_dis=results['dis']['kg'],
             var_dis=results['dis']['var'])
    print(f"\nData saved successfully to {save_path}")
except Exception as e:
    print(f"Error saving data: {e}")
```

```python
### Attinger functions ###
def attinger_amean(variance, scale, a, lc, Kg): 
    keff_scale = Kg * np.exp((variance/6)+(variance/3)*(erf((lc*a)/(np.sqrt(2)*scale)))**3)
    return keff_scale

def attinger_var(variance, scale, a, lc, Kg): #Attinger (Eq. 3.2.3, Boschan & Noetinger 2012)
    var_scale = Kg**2 * variance * (erf ((lc*a)/(np.sqrt(2)*scale)))**3
    return var_scale
```

```python
### 3D Arithmetic mean - Plot and save data ###

def calculate_arithmetic_mean(conDir, varDir, scales, nseeds, kFG, kgFG, lc):
    """ Calculate arithmetic mean statistics for different configurations. """
    results = {}
    
    for icon, rdir1 in enumerate(conDir):
        results[icon] = {}
        for ivar, rdir2 in enumerate(varDir):
            keff = np.zeros(len(scales))
            
            for iscale, scale in enumerate(scales):
                try:
                    aux = np.array([])
                    
                    if scale == scales[0]:
                        aux = np.array([kFG[icon][ivar]])
                    else:
                        for iseed in range(nseeds):
                            filepath = f"{rdir1}{rdir2}{iseed}"
                            
                            if scale == scales[-1]:
                                data = np.loadtxt(f"{filepath}/SolverRes.txt")[0]
                                aux = np.append(aux, data)
                            else:
                                data = np.load(f"{filepath}/KpostProcess/Kperm{scale}.npy")
                                aux = np.append(aux, np.mean(data.reshape(-1)))
                                
                    # Remove invalid values
                    aux = aux[~np.isnan(aux)]
                    aux = aux[aux > 0]
                    
                    if len(aux) > 0:
                        keff[iscale] = np.mean(aux)
                        
                except Exception as e:
                    print(f"Error processing scale {scale}: {e}")
                    continue
                    
            results[icon][ivar] = {
                'keff': keff,
                'normalized': keff/kgFG[icon][ivar]
            }
            
    return results

def define_reference_values():
    """Define reference values for different connectivity structures"""
    return {
        'janko': [1.52, 1.4, 1.2], 
        'matheron': [1.087, 1.181, 1.396],
        'gutjahr': [1.083, 1.167, 1.333]
    }


def plot_arithmetic_mean(scales, lc, results, conDir, varDir, varSim, color, marker, 
                        stCon, janko, matheron, varFG, kgFG, scaleAtt, ref_values):
    
    fig, ax = plt.subplots(1, 3, figsize=(16, 6), layout='constrained', sharey=True)
    
    linestyle = ['--', '-.', ':']
    
    for icon in range(len(conDir)):
        for ivar in range(len(varDir)):
            # Plot main data
            normalized_keff = results[icon][ivar]['normalized']
            ax[icon].loglog(scales/lc, normalized_keff,
                          color=color[icon], 
                          marker=marker[ivar],
                          markersize=6,
                          fillstyle='full',
                          linewidth=0.75,
                          linestyle='-',
                          label=rf'$\sigma^2_{{Y}}={varSim[ivar]:.1f}$')
            
            # Add analytical solution for intermediate connectivity
            if icon == 1:
                ax[1].loglog(scaleAtt/lc, 
                            attinger_amean(varFG[icon][ivar], scaleAtt, 1.3, lc, kgFG[icon][ivar]),
                            label=rf'Analytical: $\sigma^2_{{Y}}={varSim[ivar]:.1f}$',
                            color='black',
                            linewidth=1.25,
                            linestyle=linestyle[ivar])
            
            # Add reference points
            ax[icon].loglog(max(scales)/lc, ref_values['janko'][icon],
                          marker='x', markersize=12,
                          color='black', fillstyle='full')
            if icon == 1:
                ax[1].loglog(max(scales)/lc, ref_values['matheron'][ivar],
                            marker='D', markersize=8,
                            color='black', fillstyle='full')
                
        ax[icon].set_xlabel(r'$\lambda/I$', fontsize=20)
        if icon == 0:
            ax[icon].set_ylabel(r'$⟨K_{eq}⟩ \ / \ K_{g}$', fontsize=20)
        
        ax[icon].get_yaxis().set_major_formatter(ScalarFormatter())
        ax[icon].yaxis.set_minor_formatter(NullFormatter())
        ax[icon].set_yticks(ticks=np.arange(1,3,0.25), labels=np.arange(1,3,0.25))
        ax[icon].get_yaxis().set_major_formatter(StrMethodFormatter('{x:,.2f}'))
        ax[icon].tick_params(axis='both', labelsize=14)
        ax[icon].legend(fontsize=11)
    
    return fig, ax

def save_results(results, metadata, filename='3D_data/3D_arithmetic_mean_data.npz'):
    os.makedirs('3D_data', exist_ok=True)
    np.savez(filename,
             results=results,
             metadata=metadata)
    print(f"Saved data to {filename}")

def main():
    for directory in ['3D_plots', '3D_data']:
        os.makedirs(directory, exist_ok=True)
    
    ref_values = define_reference_values()
    
    results = calculate_arithmetic_mean(conDir, varDir, scales, nseeds, 
                                     kFG, kgFG, lc)

    fig, ax = plot_arithmetic_mean(
        scales=scales,
        lc=lc,
        results=results,
        conDir=conDir,
        varDir=varDir,
        varSim=varSim,
        color=color,
        marker=marker,
        stCon=stCon,
        janko=ref_values['janko'],
        matheron=ref_values['matheron'],
        varFG=varFG,
        kgFG=kgFG,
        scaleAtt=scaleAtt,
        ref_values=ref_values 
    )
    
    
    plt.savefig('3D_plots/3D_arithmetic_mean.png', dpi=300, bbox_inches='tight')
    plt.close()
    
    metadata = {
        'scales': scales,
        'varSim': varSim,
        'stCon': stCon,
        'reference_values': ref_values
    }
    save_results(results, metadata)
    
    print("Analysis complete. Results saved in 3D_plots/ and 3D_data/")

if __name__ == "__main__":
    main()
```

```python
### Geometric mean - Plot and save data ###

def calculate_geometric_mean(conDir, varDir, scales, nseeds, lc, kgFG):
    """ Calculate geometric mean statistics for different configurations. """
    results = {}
    
    for icon, rdir1 in enumerate(conDir):
        results[icon] = {}
        for ivar, rdir2 in enumerate(varDir):
            keff = np.zeros(len(scales))
            
            for iscale, scale in enumerate(scales):
                try:
                    aux = np.array([])
                    
                    if scale == scales[0]:
                        aux = np.array([kgFG[icon][ivar]])
                    else:
                        for iseed in range(nseeds):
                            filepath = f"{rdir1}{rdir2}{iseed}"
                            
                            if scale == scales[-1]:
                                data = np.loadtxt(f"{filepath}/SolverRes.txt")[0]
                                aux = np.append(aux, data)
                            else:
                                data = np.load(f"{filepath}/KpostProcess/Kperm{scale}.npy")
                                aux = np.append(aux, data.reshape(-1))
                
                    aux = aux[~np.isnan(aux)]
                    aux = aux[aux > 0]
                    
                    if len(aux) > 0:
                        keff[iscale] = np.exp(np.mean(np.log(aux)))
                        
                except Exception as e:
                    print(f"Error processing scale {scale}: {e}")
                    continue
                    
            results[icon][ivar] = {
                'keff': keff,
                'normalized': keff/kgFG[icon][ivar]
            }
            
    return results

def plot_geometric_mean(scales, lc, results, stCon, varSim, color, marker, linestyle):
    """Create geometric mean plot with consistent styling"""
    fig, ax = plt.subplots(1, 1, layout="constrained", figsize=(10, 8))
    
    for icon in range(len(results)):
        for ivar in range(len(results[icon])):
            normalized_keff = results[icon][ivar]['normalized']
            
            ax.loglog(scales/lc, normalized_keff,
                     label=f"{stCon[icon]}, $\\sigma^2_{{Y}}={varSim[ivar]}$",
                     color=color[icon],
                     marker=marker[ivar],
                     markersize=7,
                     fillstyle='full',
                     linewidth=1.5,
                     linestyle=linestyle[icon])
            
    ax.set_xlabel(r'$\lambda / I$', fontsize=20)
    ax.set_ylabel(r'$⟨⟨K_{eq}⟩⟩ \ / \ K_{g}$', fontsize=20)
    ax.get_yaxis().set_major_formatter(ScalarFormatter())
    ax.yaxis.set_minor_formatter(NullFormatter())
    ax.set_yticks(ticks=np.arange(0.75, 1.5, 0.25), labels=np.arange(0.75, 1.5, 0.25))
    ax.tick_params(axis='both', labelsize=14)
    ax.get_yaxis().set_major_formatter(StrMethodFormatter('{x:,.2f}'))
    ax.set_ylim(0.98, 1.75)
    ax.legend(fontsize=12, frameon=True, loc='best')
    ax.grid(True, which='both', linestyle='--', alpha=0.6)
    
    return fig, ax

def save_results(results, metadata, filename='3D_data/3D_geometric_mean_data.npz'):
    os.makedirs('3D_data', exist_ok=True)
    np.savez(filename,
             results=results,
             metadata=metadata)
    print(f"Saved data to {filename}")

def main():
    # Create output directories
    for directory in ['3D_plots', '3D_data']:
        os.makedirs(directory, exist_ok=True)
    
    results = calculate_geometric_mean(
        conDir=conDir,
        varDir=varDir,
        scales=scales,
        nseeds=nseeds,
        lc=lc,
        kgFG=kgFG
    )
    
    fig, ax = plot_geometric_mean(
        scales=scales,
        lc=lc,
        results=results,
        stCon=stCon,
        varSim=varSim,
        color=color,
        marker=marker,
        linestyle=linestyle
    )
    
    plt.savefig('3D_plots/3D_geometric_mean.png', dpi=300, bbox_inches='tight')
    plt.close()
    
    metadata = {
        'scales': scales,
        'lc': lc,
        'varSim': varSim,
        'stCon': stCon
    }
    save_results(results, metadata)

if __name__ == "__main__":
    main()
```

```python
### Probability density function - Plot and save data ###

def plot_and_save_pdfs(conDir, varDir, scalesHist, nseeds, xlim, xnum, ynum):
    """ Plot and save probability density functions with histogram data. """
    
    plots_dir = "3D_plots"
    data_dir = "3D_data"
    for directory in [plots_dir, data_dir]:
        if not os.path.exists(directory):
            os.makedirs(directory)

    scale_col = [
        ["rosybrown", "lightcoral", "indianred", "red", "sienna", "darkred"],
        ["turquoise", "springgreen", "mediumseagreen", "lime", "green", "darkslategrey"],
        ["lightblue", "lightskyblue", "deepskyblue", "steelblue", "blue", "navy"]
    ]

    num = [['a','b','c'], ['d','e','f'], ['g','h','i']]
    stCon = ['HCS', 'ICS', 'LCS']

    pdf_stats = {
        'scales': np.array(scalesHist),
        'connectivity': np.array(stCon),
        'variances': [],
        'means': [],
        'stds': [],
        'skewness': [],
        'hist_data': [] 
    }

    for ivar in range(len(varDir)):
        fig, ax = plt.subplots(1, 3, sharey=True, layout='constrained')
        fig.set_figwidth(15)
        rdir2 = varDir[ivar]
        hist_data_var = []  
        
        for icon in range(len(conDir)):
            rdir1 = conDir[icon]
            hist_data_con = []  
            
            for iscale in range(len(scalesHist)):
                scale = scalesHist[iscale]
                keff = np.array([])
                
                for iseed in range(nseeds):
                    try:
                        if scale == 1:
                            aux = np.load(rdir1+rdir2+str(iseed)+'/k.npy').reshape(-1)
                        else:
                            aux = np.load(rdir1+rdir2+str(iseed)+'/KpostProcess/Kperm'+str(scale)+'.npy').reshape(-1)
                            aux = aux[aux > 0]
                            aux = aux[~np.isnan(aux)]
                        
                        if aux.size > 0:  
                            keff = np.append(keff, aux)
                    except Exception as e:
                        print(f"Error processing seed {iseed}: {e}")
                        continue

                if keff.size > 0:  
                    log_keff = np.log(keff)
                    
                    hist_counts, hist_bins = np.histogram(log_keff, bins='scott', density=True)
                    hist_centers = (hist_bins[:-1] + hist_bins[1:]) / 2
                    
                    hist_data_con.append({
                        'scale': scale,
                        'bin_centers': hist_centers,
                        'counts': hist_counts,
                        'bins': hist_bins
                    })
                    
                    pdf_stats['variances'].append(float(np.var(log_keff)))
                    pdf_stats['means'].append(float(np.mean(log_keff)))
                    pdf_stats['stds'].append(float(np.std(log_keff)))
                    pdf_stats['skewness'].append(float(scipy.stats.skew(log_keff)))
                   
                    ax[icon].hist(log_keff, 
                                bins='scott',
                                density=True,
                                histtype='step',
                                label=f'λ={scale}',
                                linewidth=2,
                                color=scale_col[icon][iscale])

            hist_data_var.append(hist_data_con)
            
            ax[icon].set_xlabel(r'$log(K_{eq})$', fontsize=20)
            if icon == 0:
                ax[icon].set_ylabel(r'$P[log(K_{eq})]$', fontsize=20)
            ax[icon].tick_params(axis='both', labelsize=14)
            ax[icon].set_xlim(xlim[ivar])
            ax[icon].legend(fontsize=12, frameon=True, loc='center left')
            ax[icon].text(xnum[ivar], ynum[ivar], num[ivar][icon], 
                         style='normal', fontsize=18)
            ax[icon].grid(True, which='both', linestyle='--', alpha=0.6)

        pdf_stats['hist_data'].append(hist_data_var)
        
        plt.suptitle(f"Variance = {varDir[ivar].rstrip('/')}", fontsize=16)
        plt.tight_layout()
        
        plot_path = os.path.join(plots_dir, f"3D_pdf_variance_{varDir[ivar].rstrip('/')}.png")
        plt.savefig(plot_path, format='png', dpi=100, bbox_inches='tight')
        print(f"Plot saved as {plot_path}")
        plt.close()

    for key in ['variances', 'means', 'stds', 'skewness']:
        pdf_stats[key] = np.array(pdf_stats[key])

    data_path = os.path.join(data_dir, "3D_pdf_stats.npz")
    np.savez(data_path, **pdf_stats)
    print(f"Statistics and histogram data saved as {data_path}")

    hist_data_path = os.path.join(data_dir, "3D_pdf_histogram_data.npz")
    hist_data_dict = {
        'variances': varDir,
        'connectivity': stCon,
        'scales': scalesHist,
        'histogram_data': pdf_stats['hist_data']
    }
    np.savez(hist_data_path, **hist_data_dict)
    print(f"Detailed histogram data saved as {hist_data_path}")

conDir = ['./3d/3dcon/', './3d/3dint/', './3d/3ddis/']
varDir = ['0.5/', '1/', '2/']
nseeds = 64
scalesHist = [2, 4, 8, 16, 32, 64]
xlim = ([-2, 2], [-3, 3], [-4, 4])
xnum = [-1.9, -2.85, -3.8]
ynum = [2.45, 1.65, 1.175]

plot_and_save_pdfs(
    conDir=conDir,
    varDir=varDir,
    scalesHist=scalesHist,
    nseeds=nseeds,
    xlim=xlim,
    xnum=xnum,
    ynum=ynum
)
```

```python
def load_and_plot_detailed_pdfs(data_file='3D_data/3D_pdf_histogram_data.npz'):
    """ Load and plot detailed probability density functions from saved data. """
    
    # Setup colors and styles
    scale_col = [
        ["rosybrown", "lightcoral", "indianred", "red", "sienna", "darkred"],
        ["turquoise", "springgreen", "mediumseagreen", "lime", "green", "darkslategrey"],
        ["lightblue", "lightskyblue", "deepskyblue", "steelblue", "blue", "navy"]
    ]
    
    try:
        # Create output directory if it doesn't exist
        os.makedirs('3D_plots', exist_ok=True)
        
        # Load the saved data
        data = np.load(data_file, allow_pickle=True)
        variances = data['variances']
        connectivity = data['connectivity']
        scales = data['scales']
        hist_data = data['histogram_data']
        
        # Create plots for each variance value
        for ivar, variance in enumerate(variances):
            fig, axes = plt.subplots(1, 3, sharey=True, figsize=(15, 6), layout='constrained')
            
            for icon, (ax, conn_type) in enumerate(zip(axes, connectivity)):
                hist_data_conn = hist_data[ivar][icon]
                
                for iscale, scale_data in enumerate(hist_data_conn):
                    if scale_data is not None:
                        ax.plot(scale_data['bin_centers'],
                               scale_data['counts'],
                               label=f'λ={scales[iscale]}',
                               color=scale_col[icon][iscale],
                               linewidth=2)
                
                # Customize each subplot
                ax.set_xlabel(r'$log(K_{eq})$', fontsize=20)
                if icon == 0:
                    ax.set_ylabel(r'$P[log(K_{eq})]$', fontsize=20)
                ax.tick_params(axis='both', labelsize=14)
                ax.legend(fontsize=12, frameon=True, loc='best')
                ax.set_title(f'{conn_type}', fontsize=16)
           
                # Add grid
                ax.grid(True, alpha=0.6)
            
            # Set overall title and save
            variance_str = variance.rstrip('/')  # Remove trailing slash
            plt.suptitle(f'PDF for variance {variance_str}', fontsize=18)
            
            # Fix the file path formatting
            save_path = os.path.join('3D_plots', f'3D_pdf_variance_{variance_str}.png')
            plt.savefig(save_path, dpi=300, bbox_inches='tight')
            plt.close()
            print(f"Saved plot to {save_path}")
            
    except Exception as e:
        print(f"Error loading or plotting data: {e}")
        raise  # Re-raise the exception to see full traceback

# Make sure the output directory exists and call the function
os.makedirs('3D_plots', exist_ok=True)
load_and_plot_detailed_pdfs()
```
